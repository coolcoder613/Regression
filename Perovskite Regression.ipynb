```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df.dropna(axis = 1)
```

```python
df.corr()
```

```python
df = df.drop("id",1)
```

```python
df = df.drop("a_atom",1)
```

```python
df = df.drop("b1_atom",1)
```

```python
df = df.drop("b2_atom",1)
```

```python
df = df.drop("composition",1)
```

```python
df = df.drop("x_atom",1)
```

```python
X = df.drop("heat_of_formation", 1)
y = df["heat_of_formation"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Regression = LinearRegression()
```

```python
Regression.fit(X_train, y_train)
```

```python
Regression.score(X_test, y_test)
```

```python
X = df["distance_b2"].values.reshape(-1,1)
y = df["heat_of_formation"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
Regression = LinearRegression()
Regression.fit(X_train, y_train)
Regression.score(X_test, y_test)
```

```python
plt.scatter(X_test, y_test, s=10, c = "orange", alpha = 1/2)
plt.yticks(np.arange(0, 10, 1))
plt.plot(X_test, X_test*Regression.coef_ + Regression.intercept_)
plt.xlabel("distance_b2")
plt.ylabel("heat_of_formation")
plt.title("Energy of formation in hybrid perovskites")
plt.show()
```
